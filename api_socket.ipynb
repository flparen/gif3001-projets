# API socket
L'API socket a t introduit dans BSD UNIX 4.1 en 1981. L'API offre une abstraction des fonctions d'mission et de rception de donnes. 

Un **socket** est une interface de communication utilis par un processus utilisateur. Cette interface prend la forme d'un descripteur de fichier, et est utilis pour la transmission et rception de donnes.
Il existe plusieurs classes de **socket** (_domain_). Ceux qui nous intresse particulirement dans le cours sont les sockets de type **IPv4 et IPv6**.






Le tableau suivant rsume quelques fonctions de base du socket API en C 

|**Fonction**|**Description**|**client?**|**serveur?**|
|:-----------|--------------|:---------:|:----------:|
|socket()|Cre un socket|oui|oui|
|connect()|Initie une connexion (TCP)|oui|non|
|bind()|Lie le socket sur un adresse IP locale|opt|oui|
|listen()|Place le socket en mode coute pour recevoir des connexions client|non|oui|
|accept()|Accepte une connexion client|non|oui|
|read(), write(), recv(), send(), ...|Reoit et envoie des donnes|oui|oui|
|close()|Ferme le socket|oui|oui|


Les sockets 
Il existe plusieurs familles de socket

|**Famille**|**Description**|
|:---------:|:-------------:|
|socket.AF\_INET|IPv4|
|socket.AF\_INET6|IPv6|
|socket.AF\_UNIX|Communication locale|
|socket.AF\_PACKET|Utilis pour la transmission et rception de packet  la couche liaison (_paquet bas-niveau_)|
